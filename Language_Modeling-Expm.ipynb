```python
for word in words[:1]:
    for ch1, ch2 in zip(word, word[1:]):
        print(ch1, ch2)
```

```python
for word in words[:1]:
    chs = ['<S>'] + list(word) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        print(ch1, ch2)
```

```python
for word in words[:3]:
    chs = ['<S>'] + list(word) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        print(ch1, ch2)
    print("-------------")
```

```python
bd = {}
for word in words[:3]:
    chs = ['<S>'] + list(word) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        bd[bigram] = bd.get(bigram, 0) + 1
        print(ch1, ch2)
    print("-------------")
```

```python
# iterating over all the names
bd = {}
for word in words:
    chs = ['<S>'] + list(word) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        bd[bigram] = bd.get(bigram, 0) + 1
```
